```python
# multidimensional arrays
import numpy as np
import math
import pyarma as pa

# animations
from matplotlib.animation import FuncAnimation

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 18
mpl.rcParams['axes.labelsize'] = 16
mpl.rcParams['xtick.labelsize'] = 14
mpl.rcParams['ytick.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 16
plt.rcParams["figure.figsize"] = (7.5,5)
```

```python
# Reading data from binary files - 2 slits
prob0 = pa.mat()
prob0.load("probability0_2.bin")

prob1 = pa.mat()
prob1.load("probability1_2.bin")

prob2 = pa.mat()
prob2.load("probability2_2.bin")
```

```python
# Insert the edges where the probability is zero
prob0_2 = np.zeros((201,201))
prob0_2[1:-1,1:-1] = prob0

prob1_2 = np.zeros((201,201))
prob1_2[1:-1,1:-1] = prob1

prob2_2 = np.zeros((201,201))
prob2_2[1:-1,1:-1] = prob2
```

```python
# Colourmap 2D probability - t = 0 s

# Create figure
fig = plt.figure()
ax = plt.gca()

# Create a colour scale normalization according to the max z value in the first frame
norm = plt.cm.colors.Normalize(vmin=0.0, vmax=np.sqrt(prob0_2).max())

# Plot the first frame
img = ax.imshow(np.sqrt(prob0_2), extent=[0,1,0,1], cmap=plt.get_cmap("viridis"), norm=norm)

# Add a text element showing the time
time_txt = plt.text(0.95, 0.95, "t = {:.2e}".format(0.00), color="white", 
                    horizontalalignment="right", verticalalignment="top", fontsize=13)

# Axis labels
plt.xlabel("$x$")
plt.ylabel("$y$")

# Add a colourbar
cbar = fig.colorbar(img, ax=ax)
cbar.set_label(r"$\sqrt{p_{ij}^{t/\Delta t}}$")

# save plot as pdf
plt.savefig("prob0.pdf")
```

```python
# Reading data from binary files for animation - 2 slits
prob = pa.cube()
prob.load("probability_2.bin")
```

```python
# Insert the edges where the probability is zero
prob2 = pa.cube( 201, 201, 81, pa.fill.zeros)
prob2[1:199,1:199,:] = prob
```

```python
# Animation - 2 slits

# Create slice of probability values
prob_2 = np.zeros((81,201,201))
for i in range(80):
    prob_2[i] = np.array( prob2[pa.single_slice, i] ) 

# Array of time points
dt = 0.005
t_points = np.arange(0, 1+dt, dt)
t_min = t_points[0]

# Create figure
fig = plt.figure()
ax = plt.gca()

# Create a colour scale normalization according to the max probability value in the first frame
norm = mpl.cm.colors.Normalize(vmin=0.0, vmax=np.sqrt(prob_2[0]).max())

# Plot the first frame
img = ax.imshow(np.sqrt(prob_2[0]), extent=[0,1,0,1], cmap=plt.get_cmap("viridis"), norm=norm)

# Axis labels
plt.xlabel("$x$")
plt.ylabel("$y$")

# Add a colourbar
cbar = fig.colorbar(img, ax=ax)
cbar.set_label(r"$\sqrt{p_{ij}^n}$")

# Add a text element showing the time
time_txt = plt.text(0.95, 0.95, "t = {:.2e}".format(t_min), color="white", 
                    horizontalalignment="right", verticalalignment="top", fontsize=13)

# Function for updating the probability data and others for each frame
def animation(i):
    # Normalize the colour scale to the current frame
    norm = mpl.cm.colors.Normalize(vmin=0.0, vmax=np.sqrt(prob_2[i]).max())
    img.set_norm(norm)

    # Update probability data
    img.set_data(np.sqrt(prob_2[i]))

    # Update the time label
    current_time = t_min + i * dt
    time_txt.set_text("t = {:.3e}".format(current_time))

    return img

# Create and run the animation
anim = FuncAnimation(fig, animation, interval=1, frames=np.arange(0, len(np.sqrt(prob_2)), 2), repeat=False, blit=0)
plt.show()

# Save the animation
anim.save('./animation_2.mp4', writer="ffmpeg", bitrate=-1, fps=10)
```

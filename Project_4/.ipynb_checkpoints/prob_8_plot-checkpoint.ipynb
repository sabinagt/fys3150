```python
# Random initial state
#reading temperatures, average energy and magnetization per spin, 
#specific heat capacity and susceptibility per spin

T = np.loadtxt("7_rand_20.txt", usecols=0, dtype='double')

# L=20
energy_20 = np.loadtxt("7_rand_20.txt", usecols=1, dtype='double')
magn_20 = np.loadtxt("7_rand_20.txt", usecols=2, dtype='double')
Cv_20 = np.loadtxt("7_rand_20.txt", usecols=3, dtype='double')
X_20 = np.loadtxt("7_rand_20.txt", usecols=4, dtype='double')

# L=40
energy_40 = np.loadtxt("7_rand_40.txt", usecols=1, dtype='double')
magn_40 = np.loadtxt("7_rand_40.txt", usecols=2, dtype='double')
Cv_40 = np.loadtxt("7_rand_40.txt", usecols=3, dtype='double')
X_40 = np.loadtxt("7_rand_40.txt", usecols=4, dtype='double')

# L=60
energy_60 = np.loadtxt("7_rand_60.txt", usecols=1, dtype='double')
magn_60 = np.loadtxt("7_rand_60.txt", usecols=2, dtype='double')
Cv_60 = np.loadtxt("7_rand_60.txt", usecols=3, dtype='double')
X_60 = np.loadtxt("7_rand_60.txt", usecols=4, dtype='double')

# L=80
energy_80 = np.loadtxt("7_rand_80.txt", usecols=1, dtype='double')
magn_80 = np.loadtxt("7_rand_80.txt", usecols=2, dtype='double')
Cv_80 = np.loadtxt("7_rand_80.txt", usecols=3, dtype='double')
X_80 = np.loadtxt("7_rand_80.txt", usecols=4, dtype='double')

# L=100
energy_100 = np.loadtxt("7_rand_100.txt", usecols=1, dtype='double')
magn_100 = np.loadtxt("7_rand_100.txt", usecols=2, dtype='double')
Cv_100 = np.loadtxt("7_rand_100.txt", usecols=3, dtype='double')
X_100 = np.loadtxt("7_rand_100.txt", usecols=4, dtype='double')
```

```python
# Plotting energies 
plt.scatter(T, energy_20, color="blue",label="L=20", s=2)
plt.scatter(T, energy_40, color="red",label="L=40", s=2)
plt.scatter(T, energy_60, color="green",label="L=60", s=2)
plt.scatter(T, energy_80, color="yellow",label="L=80", s=2)
#plt.scatter(T, energy_100, color="magenta",label="L=100", s=2)

# add axis labels and legend
plt.xlabel("T [J/k$_B$]")
plt.ylabel(r"$\langle\epsilon\rangle$ [J]")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("Energy.pdf")
```

```python
# Plotting magnetization 
plt.scatter(T, magn_20, color="blue",label="L=20", s=2)
plt.scatter(T, magn_40, color="red",label="L=40", s=2)
plt.scatter(T, magn_60, color="green",label="L=60", s=2)
plt.scatter(T, magn_80, color="yellow",label="L=80", s=2)
#plt.scatter(T, magn_100, color="magenta",label="L=100", s=2)

# add axis labels and legend
plt.xlabel("T [J/k$_B$]")
plt.ylabel(r"$\langle$|m|$\rangle$")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("Magnetization.pdf")
```

```python
# Plotting specific heat capacity 
plt.scatter(T, Cv_20, color="blue",label="L=20", s=2)
plt.scatter(T, Cv_40, color="red",label="L=40", s=2)
plt.scatter(T, Cv_60, color="green",label="L=60", s=2)
plt.scatter(T, Cv_80, color="yellow",label="L=80", s=2)
#plt.scatter(T, Cv_100, color="magenta",label="L=100", s=2)

# add axis labels and legend
plt.xlabel("T [J/k$_B$]")
plt.ylabel("C$_v$ [k$_B$/J]")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("Heat_capacity.pdf")
```

```python
# Plotting susceptibility
plt.scatter(T, X_20, color="blue",label="L=20", s=2)
plt.scatter(T, X_40, color="red",label="L=40", s=2)
plt.scatter(T, X_60, color="green",label="L=60", s=2)
plt.scatter(T, X_80, color="yellow",label="L=80", s=2)
#plt.scatter(T, X_100, color="magenta",label="L=100", s=2)

# add axis labels and legend
plt.xlabel("T [J/k$_B$]")
plt.ylabel("$\chi$ [1/J]")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("Susceptibility.pdf")
```

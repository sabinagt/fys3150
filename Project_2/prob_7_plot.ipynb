```python
# multidimensional arrays
import numpy as np
import math

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 7.5
plt.rcParams["figure.figsize"] = (7,5)
```

```python
#reading x values file n = 10
x_10 = np.loadtxt("x_10.txt", usecols=0, dtype='double')

#insert boundary points
x_10 = np.append(x_10, 1)
```

```python
#reading eigenvectors file n = 10
eig1_10 = np.loadtxt("eigenvectors_10.txt", usecols=0, dtype='double')
eig2_10 = np.loadtxt("eigenvectors_10.txt", usecols=1, dtype='double')
eig3_10 = np.loadtxt("eigenvectors_10.txt", usecols=2, dtype='double')


#insert boundary points
eig1_10 = np.insert(eig1_10, 0,0)
eig2_10 = np.insert(eig2_10, 0,0)
eig3_10 = np.insert(eig3_10, 0,0)

eig1_10 = np.append(eig1_10, 0)
eig2_10 = np.append(eig2_10, 0)
eig3_10 = np.append(eig3_10, 0)
```

```python
#reading analytical eigenvectors file n = 10
eig1_an_10 = np.loadtxt("eigenvectors_analytical_10.txt", usecols=0, dtype='double')
eig2_an_10 = np.loadtxt("eigenvectors_analytical_10.txt", usecols=1, dtype='double')
eig3_an_10 = np.loadtxt("eigenvectors_analytical_10.txt", usecols=2, dtype='double')


#insert boundary points
eig1_an_10 = np.insert(eig1_an_10, 0,0)
eig2_an_10 = np.insert(eig2_an_10, 0,0)
eig3_an_10 = np.insert(eig3_an_10, 0,0)

eig1_an_10 = np.append(eig1_an_10, 0)
eig2_an_10 = np.append(eig2_an_10, 0)
eig3_an_10 = np.append(eig3_an_10, 0)

eig3_an_10 = np.negative(eig3_an_10)
```

```python
# plot results for n = 10

plt.plot(x_10,eig1_10,color="green",label="eigvec 1")
plt.plot(x_10,eig2_10,color="red",label="eigvec 2")
plt.plot(x_10,eig3_10,color="blue",label="eigvec 3")

plt.plot(x_10,eig1_an_10,color="black",linestyle = '--',label="analytic eigvec 1")
plt.plot(x_10,eig2_an_10,color="yellow", linestyle = '--',label="analytic eigvec 2")
plt.plot(x_10,eig3_an_10,color="magenta", linestyle = '--',label="analytic eigvec 3")

# add axis labels and legend
plt.xlabel("x")
plt.ylabel("v")
plt.legend()

# add a title 
plt.title("v(x) for n=10")

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("plot_7a.pdf")
```

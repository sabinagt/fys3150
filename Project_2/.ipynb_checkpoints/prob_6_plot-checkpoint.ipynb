```python
# multidimensional arrays
import numpy as np
import math

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 12
plt.rcParams["figure.figsize"] = (7,5)
```

```python
#reading file 
N = np.loadtxt("iterations_6a.txt", usecols=0, dtype='double')
N2 = np.loadtxt("iterations_6a.txt", usecols=1, dtype='double')
it = np.loadtxt("iterations_6a.txt", usecols=2, dtype='double')
```

```python
#reading file 
Nb = np.loadtxt("iterations_6b.txt", usecols=0, dtype='double')
itb = np.loadtxt("iterations_6b.txt", usecols=1, dtype='double')
```

```python
# plot results
plt.scatter(N,it,color="blue",label="num of iterations diagonal matrix")
plt.plot(N,N2,color="red",label="$N^2$")

# add axis labels and legend
plt.xlabel("N")
plt.ylabel("Iterations")
plt.legend()

# add a title 
plt.title("Number of iterations vs N")

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("plot_6a.pdf")

```

```python
# plot results
plt.scatter(N,it,color="blue",label="num of iterations diagonal matrix")
plt.scatter(Nb,itb,color="yellow",label="num of iterations dense matrix")
plt.plot(N,N2,color="red",label="$N^2$")

# add axis labels and legend
plt.xlabel("N")
plt.ylabel("Iterations")
plt.legend()

# add a title 
plt.title("Number of iterations vs N")

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("plot_6b.pdf")
```

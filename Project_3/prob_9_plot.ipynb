```python
# multidimensional arrays
import numpy as np
import math

# inline plots
import matplotlib.pyplot as plt
%matplotlib inline

# nicer figures
import matplotlib as mpl
mpl.rcParams['axes.titlesize'] = 18
mpl.rcParams['axes.labelsize'] = 16
mpl.rcParams['xtick.labelsize'] = 14
mpl.rcParams['ytick.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 16
plt.rcParams["figure.figsize"] = (7,5)

#3d plot
from mpl_toolkits import mplot3d
```

```python
#reading time t
t = np.loadtxt("time.txt", usecols=0, dtype='double')

#reading r_z values
r_z_no = np.loadtxt("r1_nointer.txt", usecols=2, dtype='double')
```

```python
# Plotting motion in the z direction as a function of time
plt.plot(t, r_z_no)

# add axis labels
plt.xlabel("t [$\mu$s]")
plt.ylabel("z [$\mu$m]")

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("zvst.pdf")
```

```python
#confront between omega values
T = 100/11 # period

omega_teo = math.sqrt(2*9.65/40.08) # square root of 2qVo/md^2
omega_exp = 2*math.pi/T 

print("theoretical value",omega_teo)
print("experimental value",omega_exp)
```

```python
#with interaction
#reading r values two particles
r_x_1 = np.loadtxt("r1_inter.txt", usecols=0, dtype='double')
r_y_1 = np.loadtxt("r1_inter.txt", usecols=1, dtype='double')
r_z_1 = np.loadtxt("r1_inter.txt", usecols=2, dtype='double')

r_x_2 = np.loadtxt("r2_inter.txt", usecols=0, dtype='double')
r_y_2 = np.loadtxt("r2_inter.txt", usecols=1, dtype='double')
r_z_2 = np.loadtxt("r2_inter.txt", usecols=2, dtype='double')
```

```python
# Plotting motion of two particles with interactions
plt.plot(r_x_1, r_y_1, color="blue",label="particle 1")
plt.plot(r_x_2, r_y_2, color="red",label="particle 2")

# add axis labels and legend
plt.xlabel("x [$\mu$m]")
plt.ylabel("y [$\mu$m]")
plt.legend()

# add grid
plt.grid(linestyle = '--', linewidth = 0.2)

# save plot as pdf
plt.savefig("xvsy_inter.pdf")
```

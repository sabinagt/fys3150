```python
# Estimate step size
t_tot = 100

h_1e2 = t_tot/1e2
h_1e3 = t_tot/1e3
h_1e4 = t_tot/1e4
h_1e5 = t_tot/1e5
h_1e6 = t_tot/1e6

h_5e4 = t_tot/5e4
h_2e5 = t_tot/2e5
```

```python
# Estimate error convergence rate FE
delta_max_FE_1e2 = np.max(abs_err_FE_1e2)
delta_max_FE_1e3 = np.max(abs_err_FE_1e3)
delta_max_FE_1e4 = np.max(abs_err_FE_1e4)
delta_max_FE_1e5 = np.max(abs_err_FE_1e5)
delta_max_FE_1e6 = np.max(abs_err_FE_1e6)

r_FE_1e3 = (math.log10(delta_max_FE_1e3/delta_max_FE_1e2))/(math.log10(h_1e3/h_1e2))
r_FE_1e4 = (math.log10(delta_max_FE_1e4/delta_max_FE_1e3))/(math.log10(h_1e4/h_1e3))
r_FE_1e5 = (math.log10(delta_max_FE_1e5/delta_max_FE_1e4))/(math.log10(h_1e5/h_1e4))
r_FE_1e6 = (math.log10(delta_max_FE_1e6/delta_max_FE_1e5))/(math.log10(h_1e6/h_1e5))

print("error convergence rate FE n = 10^3: ", r_FE_1e3)
print("error convergence rate FE n = 10^4: ", r_FE_1e4)
print("error convergence rate FE n = 10^5: ", r_FE_1e5)
print("error convergence rate FE n = 10^6: ", r_FE_1e6)

r_FE = (r_FE_1e4+r_FE_1e5+r_FE_1e6+r_FE_1e3)/4

print("total error convergence FE: ", r_FE)
```
